```python
import json
import os
import requests


def download_and_load_file(file_path, url):
    if not os.path.exists(file_path):
        response = requests.get(url, timeout=30)
        response.raise_for_status()
        text_data = response.text
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(text_data)

    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)

    return data

file_path = "datas/instruction-data.json"
url = (
    "https://raw.githubusercontent.com/rasbt/LLMs-from-scratch"
    "/main/ch07/01_main-chapter-code/instruction-data.json"
)

data = download_and_load_file(file_path, url)
print("ìƒ˜í”Œ ê°œìˆ˜:", len(data))
```

```python
def format_input(entry):
    instruction_text = (
        f"Below is an instruction that describes a task. "
        f"Write a response that appropriately completes the request."
        f"\n\n### Instruction:\n{entry['instruction']}"
    )

    input_text = f"\n\n### Input:\n{entry['input']}" if entry["input"] else ""

    return instruction_text + input_text
```

```python
import torch
from torch.utils.data import Dataset


class InstructionDataset(Dataset):
    def __init__(self, data, tokenizer):
        self.data = data

        # í…ìŠ¤íŠ¸ í† í°í™”
        self.encoded_texts = []
        for entry in data:
            instruction_plus_input = format_input(entry)
            response_text = f"\n\n### Response:\n{entry['output']}"
            full_text = instruction_plus_input + response_text
            self.encoded_texts.append(
                tokenizer.encode(full_text)
            )

    def __getitem__(self, index):
        return self.encoded_texts[index]

    def __len__(self):
        return len(self.data)
```

  - ì—¬ê¸°ì„œëŠ” ë” ì •êµí•œ ì ‘ê·¼ ë°©ì‹ì„ ì·¨í•˜ê³  ë°ì´í„° ë¡œë”ì— ì „ë‹¬í•  ìˆ˜ ìˆëŠ” ì‚¬ìš©ì ì§€ì • ì½œë ˆì´íŠ¸ í•¨ìˆ˜ë¥¼ ê°œë°œí•©ë‹ˆë‹¤.
  - ì´ ì‚¬ìš©ì ì§€ì • ì½œë ˆì´íŠ¸ í•¨ìˆ˜ëŠ” ê° ë°°ì¹˜ì˜ í›ˆë ¨ ìƒ˜í”Œì„ ë™ì¼í•œ ê¸¸ì´ë¡œ íŒ¨ë”©í•©ë‹ˆë‹¤(ë‹¨, ë°°ì¹˜ë§ˆë‹¤ ê¸¸ì´ê°€ ë‹¤ë¥¼ ìˆ˜ ìˆìŒ).

```python
def custom_collate_draft_1(
    batch,
    pad_token_id=50256,
    device="cpu"
):
    # ë°°ì¹˜ì—ì„œ ê°€ì¥ ê¸´ ì‹œí€€ìŠ¤ ì°¾ê¸°
    # ê·¸ë¦¬ê³  ìµœëŒ€ ê¸¸ì´ë¥¼ +1ì”© ì¦ê°€ì‹œì¼œ ì•„ë˜ì—ì„œ íŒ¨ë”© í† í°ì„ í•˜ë‚˜ ì¶”ê°€í•©ë‹ˆë‹¤.
    batch_max_length = max(len(item)+1 for item in batch)

    # ì…ë ¥ íŒ¨ë”© ë° ì¤€ë¹„
    inputs_lst = []

    for item in batch:
        new_item = item.copy()
        # <|endoftext|> í† í° ì¶”ê°€
        new_item += [pad_token_id]
        # batch_max_lengthê¹Œì§€ ì‹œí€€ìŠ¤ íŒ¨ë”©
        padded = (
            new_item + [pad_token_id] *
            (batch_max_length - len(new_item))
        )
        # padded[:-1]ë¥¼ í†µí•´ batch_max_lengthì˜ +1 ì„¤ì •ì„ í†µí•´ ì¶”ê°€ëœ
        # ì¶”ê°€ íŒ¨ë”© í† í°ì„ ì œê±°í•©ë‹ˆë‹¤.
        # (ì¶”ê°€ íŒ¨ë”© í† í°ì€ ì´í›„ ì½”ë“œì—ì„œ ê´€ë ¨ì´ ìˆìŠµë‹ˆë‹¤)
        inputs = torch.tensor(padded[:-1])
        inputs_lst.append(inputs)

    # ì…ë ¥ ë¦¬ìŠ¤íŠ¸ë¥¼ í…ì„œë¡œ ë³€í™˜í•˜ê³  íƒ€ê¹ƒ ì¥ì¹˜ë¡œ ì „ì†¡
    inputs_tensor = torch.stack(inputs_lst).to(device)
    return inputs_tensor
```

```python
inputs_1 = [0, 1, 2, 3, 4]
inputs_2 = [5, 6]
inputs_3 = [7, 8, 9]

batch = (
    inputs_1,
    inputs_2,
    inputs_3
)

print(custom_collate_draft_1(batch))
```

- ìœ„ì—ì„œëŠ” LLMì— ëŒ€í•œ ì…ë ¥ë§Œ ë°˜í™˜í–ˆìŠµë‹ˆë‹¤. ê·¸ëŸ¬ë‚˜ LLM í›ˆë ¨ì„ ìœ„í•´ì„œëŠ” íƒ€ê¹ƒ ê°’ë„ í•„ìš”í•©ë‹ˆë‹¤.
- LLM ì‚¬ì „ í›ˆë ¨ê³¼ ìœ ì‚¬í•˜ê²Œ, íƒ€ê¹ƒì€ ì…ë ¥ì„ ì˜¤ë¥¸ìª½ìœ¼ë¡œ 1 ìœ„ì¹˜ì”© ì´ë™í•œ ê²ƒì´ë¯€ë¡œ LLMì€ ë‹¤ìŒ í† í°ì„ ì˜ˆì¸¡í•˜ëŠ” ë°©ë²•ì„ í•™ìŠµí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
def custom_collate_draft_2(
    batch,
    pad_token_id=50256,
    device="cpu"
):
    # ë°°ì¹˜ì—ì„œ ê°€ì¥ ê¸´ ì‹œí€€ìŠ¤ ì°¾ê¸°
    batch_max_length = max(len(item)+1 for item in batch)

    # ì…ë ¥ ë° íƒ€ê¹ƒ ì¤€ë¹„
    inputs_lst, targets_lst = [], []

    for item in batch:
        new_item = item.copy()
        # <|endoftext|> í† í° ì¶”ê°€
        new_item += [pad_token_id]
        # ì‹œí€€ìŠ¤ë¥¼ max_lengthê¹Œì§€ íŒ¨ë”©
        padded = (
            new_item + [pad_token_id] *
            (batch_max_length - len(new_item))
        )
        inputs = torch.tensor(padded[:-1])  # ì…ë ¥ì„ ìœ„í•´ ë§ˆì§€ë§‰ í† í° ìë¥´ê¸°
        targets = torch.tensor(padded[1:])  # íƒ€ê¹ƒì„ ìœ„í•´ ì˜¤ë¥¸ìª½ìœ¼ë¡œ +1 ì´ë™
        inputs_lst.append(inputs)
        targets_lst.append(targets)

    # ì…ë ¥ ë¦¬ìŠ¤íŠ¸ë¥¼ í…ì„œë¡œ ë³€í™˜í•˜ê³  íƒ€ê¹ƒ ì¥ì¹˜ë¡œ ì „ì†¡
    inputs_tensor = torch.stack(inputs_lst).to(device)
    targets_tensor = torch.stack(targets_lst).to(device)
    return inputs_tensor, targets_tensor
```

- ë‹¤ìŒìœ¼ë¡œ, ëª¨ë“  íŒ¨ë”© í† í° IDë¥¼ `ignore_index`ì— ì§€ì •í•œ ê°’ìœ¼ë¡œ ë°”ê¿‰ë‹ˆë‹¤. ì´ `ignore_index`ì˜ ëª©ì ì€ ì†ì‹¤ í•¨ìˆ˜ì—ì„œ íŒ¨ë”© ê°’ì„ ë¬´ì‹œí•  ìˆ˜ ìˆë„ë¡ í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤(ìì„¸í•œ ë‚´ìš©ì€ ë‚˜ì¤‘ì— ì„¤ëª…).

- êµ¬ì²´ì ìœ¼ë¡œ, ì•„ë˜ ê·¸ë¦¼ê³¼ ê°™ì´ `50256`ì— í•´ë‹¹í•˜ëŠ” í† í° IDë¥¼ `-100`ìœ¼ë¡œ ë°”ê¿‰ë‹ˆë‹¤.


<img src="images/llm_from_scratch/ch07_compressed/12.webp" width=800px>


- (ë˜í•œ, ìƒ˜í”Œì˜ ê¸¸ì´ë¥¼ ì œí•œí•˜ê³  ì‹¶ì€ ê²½ìš°ë¥¼ ìœ„í•´ `allowed_max_length`ë¥¼ ì‚¬ìš©í–ˆìŠµë‹ˆë‹¤. GPT-2 ëª¨ë¸ì—ì„œ ì§€ì›í•˜ëŠ” 1024 í† í° ë¬¸ë§¥ í¬ê¸°ë³´ë‹¤ ê¸´ ë°ì´í„°ì…‹ìœ¼ë¡œ ì‘ì—…í•˜ë ¤ëŠ” ê²½ìš° ìœ ìš©í•©ë‹ˆë‹¤.)


```python
def custom_collate_fn(
    batch,
    pad_token_id=50256,
    ignore_index=-100,
    allowed_max_length=None,
    device="cpu"
):
    # ë°°ì¹˜ì—ì„œ ê°€ì¥ ê¸´ ì‹œí€€ìŠ¤ ì°¾ê¸°
    batch_max_length = max(len(item)+1 for item in batch)

    # ì…ë ¥ê³¼ íƒ€ê¹ƒ íŒ¨ë”© ë° ì¤€ë¹„
    inputs_lst, targets_lst = [], []

    for item in batch:
        new_item = item.copy()
        # <|endoftext|> í† í° ì¶”ê°€
        new_item += [pad_token_id]
        # ì‹œí€€ìŠ¤ë¥¼ max_lengthê¹Œì§€ íŒ¨ë”©
        padded = (
            new_item + [pad_token_id] *
            (batch_max_length - len(new_item))
        )
        inputs = torch.tensor(padded[:-1])  # ì…ë ¥ì„ ìœ„í•´ ë§ˆì§€ë§‰ í† í° ìë¥´ê¸°
        targets = torch.tensor(padded[1:])  # ëª©í‘œë¥¼ ìœ„í•´ ì˜¤ë¥¸ìª½ìœ¼ë¡œ +1 ì´ë™

        # ìƒˆë¡œ ì¶”ê°€: ëª©í‘œì—ì„œ ì²« ë²ˆì§¸ íŒ¨ë”© í† í°ì„ ì œì™¸í•œ ëª¨ë“  í† í°ì„ ignore_indexë¡œ ë°”ê¾¸ê¸°
        mask = targets == pad_token_id
        indices = torch.nonzero(mask).squeeze()
        if indices.numel() > 1:
            targets[indices[1:]] = ignore_index

        # ìƒˆë¡œ ì¶”ê°€: ìµœëŒ€ ì‹œí€€ìŠ¤ ê¸¸ì´ë¡œ ìë¥´ê¸° (ì„ íƒ ì‚¬í•­)
        if allowed_max_length is not None:
            inputs = inputs[:allowed_max_length]
            targets = targets[:allowed_max_length]

        inputs_lst.append(inputs)
        targets_lst.append(targets)

    # ì…ë ¥ ë° íƒ€ê¹ƒ ë¦¬ìŠ¤íŠ¸ë¥¼ í…ì„œë¡œ ë³€í™˜í•˜ê³  íƒ€ê¹ƒ ì¥ì¹˜ë¡œ ì „ì†¡
    inputs_tensor = torch.stack(inputs_lst).to(device)
    targets_tensor = torch.stack(targets_lst).to(device)

    return inputs_tensor, targets_tensor
```

```python
inputs, targets = custom_collate_fn(batch)
print(inputs)
print(targets)
```
